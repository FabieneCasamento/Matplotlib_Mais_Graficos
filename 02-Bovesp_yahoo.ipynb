# Carteira de ativos

```python
df_carteira = pd.read_excel("Carteira.xlsx")
display(df_carteira)
```

```python
lista_ativos = list(df_carteira["Ativos"].astype(str) + ".SA")
print(lista_ativos)
```

```python

df_cotacoes_0 = pdr.get_data_yahoo(lista_ativos, data_inicial, data_final)
display(df_cotacoes_0)
```

```python

df_cotacoes = pdr.get_data_yahoo(lista_ativos, data_inicial, data_final)["Adj Close"]
display(df_cotacoes)
```

# Ajustando os dados

```python
#para preencher com o valor do dado anterior


df_cotacoes = df_cotacoes.ffill()
print(df_cotacoes.info())
```

## Para compararmos oscilações, precisamos normalizar

```python
# normalizar os dados no mesmo valor => cair em porcentagens
#dividir todos pela primeira linha da tabela
df_cotacoes_norm = df_cotacoes / df_cotacoes.iloc[0]
df_cotacoes_norm
```

```python
df_cotacoes_norm.plot(figsize=(15, 5))
plt.legend(loc="upper left")
plt.show()
```

# Valores investidos 

```python
## Vendo valores investido

df_valor_investido = pd.DataFrame()
#df_valor_investido['BBDC4']=1
for ativo in df_carteira['Ativos']:
    #qtde_acoes = df_carteira.loc[linha,coluna] e .values[0] (para pegar apenas o valor)
    qtde_acoes = df_carteira.loc[df_carteira['Ativos']==ativo,'Qtde'].values[0]
   # print(ativo,str(qtde_acoes))
    print(ativo,qtde_acoes)
    df_valor_investido[ativo] = qtde_acoes* df_cotacoes[f"{ativo}.SA"]
   

#somar as colunas, na linha
#somar uma coluna toda ou seja todas as linhas
#df_valor_investido['Total'] = df_valor_investido.sum()
#df_valor_investido['Total'] = df_valor_investido.sum(axis=0)
#somar axis=1 varias colunas em uma unica linha
df_valor_investido['Total'] = df_valor_investido.sum(axis=1)

display(df_valor_investido)
```

## Redendo mais o ibov ou não 
# comparando carteira com o índice

```python

df_ibov = pdr.get_data_yahoo("^BVSP", data_inicial, data_final)["Adj Close"]
display(df_ibov)
```

```python
df_ibov_norm = df_ibov/ df_ibov.iloc[0]

df_valor_investido_norm = df_valor_investido/df_valor_investido.iloc[0]

```

```python
display(df_ibov_norm)
```

```python
# Gráfico que compara
df_valor_investido_norm['Total'].plot(label='Carteira',figsize=(15,5))
df_ibov_norm.plot(label='IBOVESPA')

plt.legend(fontsize=20)

plt.show()
```

## calculando retorno e correlação¶

```python
display(df_ibov_norm)

```

```python
#retorno
retorno_ibov = df_ibov_norm[-1] -1 
retorno_carteira = df_valor_investido_norm["Total"][-1] - 1

print(f"Retorno IBOV :{retorno_ibov:2f}")
print(f"Retorno IBOV :{retorno_ibov:.2f}")
print(f"Retorno IBOV :{retorno_ibov:.2%}")
print(f"Retorno Carteira :{retorno_carteira:.2%}")
```
